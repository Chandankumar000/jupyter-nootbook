### fillna to fill missing values using different ways
### interpolate to make a guess using different ways
### fillna to fill missing values using different ways

```python
type(pf.Day[0])
```

```python
pf = pd.read_excel('MissingData.xlsx' , parse_dates=["Day"])
pf.set_index('Day' , inplace=True)
pf
```

```python
new_pf=pf.fillna(0)
new_pf
```

```python
new_pf = pf.fillna({
    'Temp' : 0,
    'WindSpeed' :0,
    'Event' : 'NoEvent'
})
new_pf
```

```python
new_pf = pf.fillna(method="ffill") #values fill previous days values because its look better understanding
new_pf
```

```python
new_pf = pf.fillna(method="bfill") #values fill lastest days values because its look better understanding
new_pf
```

```python
new_pf = pf.fillna(method="bfill" , limit =1) #Limit means you fill only one missing values at a times
new_pf
```

```python
new_pf = pf.interpolate()  #for the better guess you use interpolated method
new_pf
```

```python
new_pf = pf.interpolate(method="time")  #for previous guess time dec more then date that's why use time method
new_pf
```

```python
new_pf = pf.dropna() # drop all dataframe which contains NAN values in any column
new_pf
```

```python
new_pf = pf.dropna(how="all") # drop which data where all rows contains NAN values only
new_pf
```

```python
#if you want to insert some missing data
dt = pd.date_range("01-01-2017","01-11-2017")
idx = pd.DatetimeIndex(dt)
pf = pf.reindex(idx)
pf
```

# Merge Dataframes

```python
df2 =pd.DataFrame({
    "city" : ['new york','chichago','orlando'],
    "temp" : [32,60,78],
})
df2
```

```python
df3 =pd.DataFrame({
    "city" : ['new york','chichago','orlando'],
    "temp" : [21,14,30],
})
df3
```

```python
df4 = pd.merge(df2,df3,on="city")
df4
```

```python
df5 =pd.DataFrame({
    "city" : ['new york','chichago','batlimore'],
    "temp" : [21,14,30],
})
df5
```

```python
df6 =pd.DataFrame({
    "city" : ['new york','chichago','orlando'],
    "temp" : [32,60,78],
})
df6
```

```python
df7 = pd.merge(df5,df6,on="city") #that's means this is called inner join b/w two set
df7
```

```python
# this is called union
df7 = pd.merge(df5,df6,on="city" , how="outer") 
df7
```

```python
# this is called left join
df7 = pd.merge(df5,df6,on="city" , how="left") 
df7
```
